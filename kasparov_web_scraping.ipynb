## Scraping TWICS

Download every single `.pgn` file from `https://theweekinchess.com/twic`

```python
from __future__ import print_function, division
import requests
requests.__path__
```

```python
url = 'https://theweekinchess.com/twic'
response = requests.get(url)
response.status_code
```

```python
print(response.text)
```

```python
page = response.text

soup = BeautifulSoup(page, 'lxml')
```

```python
print(soup.prettify())
```

```python
import urllib3
import time
from selenium import webdriver
import chromedriver_binary
from selenium.webdriver.common.keys import Keys

%load_ext dotenv
%dotenv
import os
```

```python
driver = webdriver.Chrome()
links = soup.find_all('a')

for link in links:
    href = link['href']

    if 'https://www.theweekinchess.com/zips/twic' in href and 'g.zip' in href:
        driver.get(href)
```
